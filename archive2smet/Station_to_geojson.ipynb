```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# 1) Read your CSV
csv_file = "/home/benjamin/Documents/Doc/subgridding_hrdps/ancillary_data_HRDPS_old/Grid_creation/HRDPS_old_20x20_voronoi_utm_attributes.csv"
df = pd.read_csv(csv_file)

# Make sure your CSV has columns like: 'name', 'lat', 'lon'
# If your column names differ, rename them:
df = df.rename(columns={'station_name': 'station_id', 'latitude': 'latitude', 'longitude': 'longitude'})

# 2) Create geometry column
df['geometry'] = df.apply(lambda row: Point(row['longitude'], row['latitude']), axis=1)

# 3) Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# 4) Keep only the needed columns
gdf = gdf[['station_id', 'geometry']]

# 5) Export to GeoJSON
geojson_file = "/home/benjamin/Documents/Doc/SHorton_HRDPS/nwp_archives/archive2smet/stations.geojson"
gdf.to_file(geojson_file, driver='GeoJSON')

print(f"GeoJSON saved to {geojson_file}")
```
